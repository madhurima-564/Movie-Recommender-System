```python
import numpy as np
import pandas as pd
import ast
```

```python
mov = pd.read_csv('/content/tmdb_5000_movies.csv')
cre = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
mov.head()
```

```python
cre.head()
```

```python
mov = mov.merge(cre,on='title')
```

```python
mov.head()
```

```python
mov.shape
```

```python
mov.info()
```

```python
mov= mov[['movie_id', 'title', 'overview', 'genres','keywords', 'cast', 'crew']]
```

```python
mov.head()
```

```python
mov.info()
```

```python
mov.isnull().sum()
```

```python
mov.dropna(inplace = True)
```

```python
mov.duplicated().sum()
```

```python
mov.iloc[0].genres
```

```python
def convert(obj):
  L = []

  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
mov['genres']= mov['genres'].apply(convert)
```

```python
mov.head()
```

```python
mov['keywords']= mov['keywords'].apply(convert)
```

```python
mov.head()
```

```python
def convert2(obj):
  L = []
  counter = 0

  for i in ast.literal_eval(obj):
    if counter != 4:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L
```

```python
mov['cast'] = mov['cast'].apply(convert2)
```

```python
mov.head()
```

```python
def fetch_d(obj):
  L = []

  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
mov['crew'] = mov['crew'].apply(fetch_d)
```

```python
mov.head()
```

```python
mov['overview'] = mov['overview'].apply(lambda x:x.split())
```

```python
mov.head()
```

```python
mov['genres'] = mov['genres'].apply(lambda x:[i.replace(" ","") for i in x])
mov['keywords'] = mov['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
mov['cast'] = mov['cast'].apply(lambda x:[i.replace(" ","") for i in x])
mov['crew'] = mov['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
mov.head()
```

```python
mov['tags'] = mov['overview'] + mov['genres'] + mov['keywords'] + mov['cast'] + mov['crew']
```

```python
mov.head()
```

```python
movies = mov[['movie_id', 'title', 'tags']]
```

```python
movies['tags']= movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

```python
movies.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies['tags'][1300]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(sim[0])), reverse = True, key = lambda x:x[1])
```

```python
def recommend(movie):
  movie_index = movies[movies['title'] == movie].index[0]
  distances = sim[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
